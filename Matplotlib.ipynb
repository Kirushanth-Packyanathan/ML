```python
# import matplotlib library
import matplotlib.pyplot as plt

import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)

print(x)
print(y)
print(z)
```

**Plotting the Data**

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin value')
plt.title('sin wave')
plt.show()
```

```python
# parabole
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

**Bar Plot**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','German']
people = [100,50,150,40]
ax.bar(languages, people)
plt.xlabel('Languages')
plt.ylabel('Number of people')
plt.title('Bar Plot')
plt.show()
```

**Pie Chart**

```python
 fig1 = plt.figure()
 ax1 = fig1.add_axes([0,0,1,1])
 languages = ['English','French','Spanish','German']
 people = [100,50,150,40]
 ax1.pie(people,labels=languages,autopct='%1.2f%%')
 plt.show()
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)

#Creating the Figure to hold the plot
figure2=plt.figure()

#Creating Axes:[0, 0] is the lower-left corner, and [1, 1] defines width and height
axes2=figure2.add_axes([0,0,1,1])

#Plotting
axes2.scatter(x,y,color='r')


axes2.set_xlabel('x')
axes2.set_title('scatter plot')
axes2.scatter(x,z,color='b')
plt.show()
```

3D Scatter Plot

```python
#Creating the Figure to hold the plot
figure3=plt.figure()

#Creating Axes
ax=plt.axes(projection='3d')

# z is an array of 100 random values between 0 and 20, generated using np.random.random(). Multiplying by 20 scales the random values.
z=20*np.random.random(100)

x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
