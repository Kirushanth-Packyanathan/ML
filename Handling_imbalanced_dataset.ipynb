**Imbalanced dataset**

A dataset with an unequal class distribution

```python
import numpy as np
import pandas as pd
```

```python
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data = credit_card_data.dropna(how='any')
```

```python
credit_card_data.isnull().sum()
```

```python
# determine the distribution of two classes
credit_card_data['Class'].value_counts()
```

This is highly imbalanced dataset

0  ---> Legit transaction
1  ---> Fraudulent Transaction

```python
# separating the legit and fradulent transaction
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

**`Under Sampling`**

Building sample dataset containing similar distribution of legit and fradulent transactions

```python
legit_sample = legit.sample(n=88)
```

```python
print(legit_sample.shape)
print(fraud.shape)
```

Concatinate two data frames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```
